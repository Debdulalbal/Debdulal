```python
#zomato data analysis project

#step-1:Importing libraries


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#step-2:creating data frame

dataframe=pd.read_csv("/content/Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

```python
#convert the data type of coloum 'RATE'

def handlerate(value):
  value=str(value).split('/');
  value=value[0];
  return float(value);

dataframe['rate']=dataframe['rate'].apply(handlerate);
print(dataframe.head())
```

```python
dataframe.info()
```

```python
# type of resturant

dataframe.head()
```

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes': grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel(" Resturant's",c="blue",size=20)
plt.ylabel("Votes",c="blue",size=20)
plt.title("Types of resturant",c="red",size=20)
```

```python
#conclusion -dining resturant has receive maximum votes
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("Rate Distribution")
plt.xlabel("Rate",c="blue")
plt.ylabel("Frequency",c="blue")
plt.show()
```

```python
#conclusion - maximum resturant receive rating from 3.4 to 4
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=dataframe,y='rate',x='online_order')
```

```python
#conclusion-online mode receive more rating than offline mode
```

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order',fill_value=0,aggfunc='size')
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu',fmt='d')
plt.title('heatmap')
plt.xlabel("Online_order")
plt.ylabel("Listed_in(type)")

plt.show()
```
